```python
datasets = [df_1, df_2, df_3, df_4, df_5, df_6, df_7]
merged_datasets = pd.concat(datasets, axis=1, keys=range(1, len(datasets) + 1))
merged_datasets.columns = merged_datasets.columns.map('{0[0]}_{0[1]}'.format)
```

```python
merged_datasets
```

```python
column_names = merged_datasets.columns
print(column_names)
```

```python
df_combined = merged_datasets[['7_All States']].copy()
df_combined.columns = df_combined.columns.str.strip().str.replace('[^\w\s]', '').str.replace('\s+', '_')

# Replace the first NaN value with 'Hawaii' in '7_All States' column
df_combined['7_All States'].fillna('Hawaii', limit=1, inplace=True)

df_combined.drop_duplicates(subset='7_All States', keep='first', inplace=True)

columns_to_combine = ['1_Good Air Quality Days', '1_Moderate Air Quality Days', '1_Unhealthy Air Quality Days', '2_Rank', '2_Final\nScore', '2_Outcome Factor Rank (1st = Best)', '2_Cost Factor Rank (1st = Lowest)', '2_Access Factor Rank (1st = Best)', '3_Rank', '3_Value', '4_Rank', '4_Value', '6_Rank', '6_Final\nScore', '6_Outcome Factor Rank (1st = Best)', '6_Cost Factor Rank (1st = Lowest)', '6_Access Factor Rank (1st = Best)', '7_Value']

df_combined = pd.concat([df_combined] + [merged_datasets[column] for column in columns_to_combine], axis=1)
df_combined

```

```python
df_combined.dropna(subset=['7_All States'], inplace=True)
df_combined
```

```python
# List of columns to process
columns_to_process = ['4_Value', '3_Value', '1_Good Air Quality Days', '1_Moderate Air Quality Days', '1_Unhealthy Air Quality Days']

# Iterate over the columns and apply the desired operations
for column in columns_to_process:
    # Check if the column is of string type
    if df_combined[column].dtype == 'object':
        # Remove the percentage symbol from the values
        df_combined[column] = df_combined[column].str.replace('%', '')
        # Convert the column to float type
        df_combined[column] = df_combined[column].astype(float)

    # Forward fill (ffill) - Replace NaN values with the most recent non-missing value
    df_combined[column].fillna(method='ffill', inplace=True)
    # Forward interpolation - Estimate missing values based on the next non-missing value
    df_combined[column].interpolate(method='linear', inplace=True)

df_combined
```

```python

# List of columns to process
columns_to_process = ['7_Value', '2_Rank', '2_Final\nScore', '2_Outcome Factor Rank (1st = Best)', '2_Cost Factor Rank (1st = Lowest)', '2_Access Factor Rank (1st = Best)', '3_Rank', '4_Rank', '6_Rank']

# Iterate over the columns and apply the desired operations
for column in columns_to_process:
    # Check if the column is of string type
    if df_combined[column].dtype == 'object':
        # Remove any unwanted characters from the values (e.g., newline characters)
        df_combined[column] = df_combined[column].str.replace('\n', '')

    # Convert the column to float type
    df_combined[column] = df_combined[column].astype(float)

    # Forward fill (ffill) - Replace NaN values with the most recent non-missing value
    df_combined[column].fillna(method='ffill', inplace=True)

    # Interpolation - Estimate missing values based on surrounding data points
    df_combined[column].interpolate(method='linear', inplace=True)

df_combined
```

```python
# List of columns to drop
columns_to_drop = ['6_Outcome Factor Rank (1st = Best)', '6_Cost Factor Rank (1st = Lowest)', '6_Access Factor Rank (1st = Best)', '7_Value']

# Drop the columns from the DataFrame
df_combined.drop(columns_to_drop, axis=1, inplace=True)
df_combined
```

```python
# Rename the columns
df_combined.rename(columns={'3_Value': '2022 overall score', '4_Value': '2023 overall score', '3_Rank': '2022 Overall Ranking', '4_Rank':'2023 Overall Ranking', '6_Rank':'Overall Current Ranking(Highter is the best)'}, inplace=True)
df_combined
```
